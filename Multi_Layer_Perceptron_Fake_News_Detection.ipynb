# Multi-Layer Perceptron

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#import data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cleaned_data.csv')
df
```

```python
# spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['cleaned_title'], df['label'], test_size=0.2, random_state=42)
```

```python
# creating TF-IDF vectorizer
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

### Hidden Layer Configurations

1.   One Hidden Layer (2,)
2.   One Hidden Layer (10,)
3.   Two Hidden Layer (10,10,)
4.   One Hidden Layer (100,)
5.   Two Hidden Layer (100,100,)



### **Configuration 1: One Hidden Layer (2,)**

```python
# initializing and training MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(2,), max_iter=1000, random_state=42)
mlp.fit(X_train_tfidf, y_train)
```

```python
# predicting labels for the test set
y_pred = mlp.predict(X_test_tfidf)
```

```python
# initializing and training MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=42)
mlp.fit(X_train_tfidf, y_train)
# predicting labels for the test set
y_pred = mlp.predict(X_test_tfidf)
# calculating & printing accuracy
accuracy_2 = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
loss_values_2 = mlp.loss_curve_
plt.plot(loss_values_2)
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```python
# initializing and training MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(10,10,), max_iter=1000, random_state=42)
mlp.fit(X_train_tfidf, y_train)
# predicting labels for the test set
y_pred = mlp.predict(X_test_tfidf)
# calculating & printing accuracy
accuracy_3 = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
loss_values_3 = mlp.loss_curve_
plt.plot(loss_values_3)
plt.show()
```

```python
# initializing and training MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
mlp.fit(X_train_tfidf, y_train)
# predicting labels for the test set
y_pred = mlp.predict(X_test_tfidf)
# calculating & printing accuracy
accuracy_4 = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
loss_values_4 = mlp.loss_curve_
plt.plot(loss_values_4)
plt.show()
```

```python
# initializing and training MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(100,100,), max_iter=1000, random_state=42)
mlp.fit(X_train_tfidf, y_train)
# predicting labels for the test set
y_pred = mlp.predict(X_test_tfidf)
# calculating & printing accuracy
accuracy_5 = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
loss_values_5 = mlp.loss_curve_
plt.plot(loss_values_5)
plt.show()
```
